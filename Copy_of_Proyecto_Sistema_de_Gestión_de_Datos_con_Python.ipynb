# **PROYECTO: SISTEMA DE GESTIÓN DE DATOS CON PYTHON**

Una vez visto el ***Módulo 1: Fundamentos de Programación***, se presenta la siguiente actividad para que el alumnado pueda poner a prueba lo aprendido durante este módulo.

Dentro de la siguiente dinámica, realizar un **menú CRUD** donde el tema a emplear es libre. Cada función deberá ejecutar una **operación del CRUD**, agregando una opción de **visualización**:

*   **Ver todos los registros**
*   **Crear nuevo registro**
*   **Consultar registro**
*   **Actualizar registro**
*   **Eliminar registro**
*   **Salir menú**

```python
# Utils

# Función para evaluar vacios
def is_empty(value):
  if value is None or value == "":
    return True
  elif isinstance(value, str):
    if value.strip() == '':
      return True
  elif isinstance(value, list):
    return len(value) == 0
  elif isinstance(value, dict):
    return len(value.keys()) == 0
  else:
    return False

# Obtiene el tamaño maximo de caracteres de una lista dado una clave
def get_max_size(_list, _key):
  newList = [len(str(item[_key])) for item in _list] if isinstance(_list, list) else [0]
  return max(newList)

# Valida tipo de dato de cada hijo de la lista
def is_children_type(_list, _type):
  isChildType = True
  for child in _list:
    if not isinstance(child, _type):
      isChildType = False
      break
  return isChildType

# Imprime todos los registros en formato de tabla
def view_table(_list, headers):
  if (not isinstance(_list, list)):
    print('1° argumento debe de ser lista.')
    return
  if (not isinstance(_list, list)):
    print('2° argumento debe de ser lista.')
    return
  if not is_children_type(_list, dict):
    print('Los elementos de la lista deben ser de tipo diccionario.')
    return

  headerRow = [{ column['col']: column['label'].replace(' ', '_') for column in headers }]
  maxSizes = {}
  headersStr = ''
  rows = ''
  rowColumn = 'row'
  listLength = len(_list)
  listLength = len(str(listLength)) if len(str(listLength)) > len(rowColumn) else len(rowColumn)

  headersStr += f'{rowColumn.capitalize()} {(listLength - len(rowColumn)) * ' '}'

  # Crea la cabecera de la tabla en string
  # Agrega espacios calculados a la derecha
  for item in headers:
    column = item['col']
    label = item['label'].replace(' ', '_')
    maxSizes[column] = get_max_size(_list + headerRow, column)
    headersStr += f'{label.title()} {(maxSizes[column] - len(label)) * ' '}'

  # Crear las filas de la tabla en string
  # Agrega espacios calculados a la derecha
  index = 1
  columnIndex = 0
  for record in _list:
    for item in headers:
      column = item['col']
      if columnIndex == 0:
        rows += f'{index} {(listLength - len(str(index))) * ' '}'
      rows += f'{record[column]} {(maxSizes[column] - len(str(record[column]))) * ' '}'
      columnIndex += 1
    rows += '\n'
    index += 1
    columnIndex = 0

  print(headersStr)
  if is_empty(_list):
    print('Sin registros por el momento')
  else:
    print(rows)

# Buscar dentro de una lista
def search(_list = [], headers = [], search = '', columns = None):
  if is_empty(_list):
    print('1° argumento "Lista" esta vacía\n')
    return
  elif is_empty(search):
    print('2° argumento "buscar" esta vacío\n')
    return
  elif not isinstance(_list, list):
    print('1° argumento debe ser una lista\n')
    return
  elif not isinstance(search, str):
    print('2° argumento debe ser una cadena\n')
    return
  if not isinstance(columns, list):
    columns = []
    if is_children_type(_list, dict):
      columns = [column['col'] for column in headers]

  result = []
  search = search.lower().strip()
  foundCount = 0
  for row in _list:
    for col in columns:
      text = str(row[col])
      text = text.lower().strip()
      if text.find(search) != -1:
        foundCount += 1
    if foundCount > 0:
      result.append(row)
      foundCount = 0

  print('\nRegistros encontrados: ', len(result))
  view_table(result, headers)
  if len(result) == 0:
    print('No se encontraron resultados\n')

# Encontrar registro por su clave y valor
def find_by(_list, key, value):
  itemFound = None
  index = 0
  for item in _list:
    if item[key] == value:
      itemFound = item
      break
    index += 1
  return {
      "record": itemFound,
      "index": index
  }
```

```python
# Menu de alumnos

def show_students_menu():
  print("Base de datos de alumnos (Master's degree in data science and AI):\n"+
        '1. Ver todos los alumnos\n'+
        '2. Crear nuevo alumno\n'+
        '3. Consultar alumno\n'+
        '4. Actualizar alumno\n'+
        '5. Eliminar alumno\n'+
        '6. Salir del menú\n')
```

```python
# Consultar alumnos

def init_find_students():
  if is_empty(dataBase):
    print('La lista no puede ser vacía.')
    return
  if not isinstance(dataBase, list):
    print('El argumento debe de ser una lista.')
    return

  exit = False
  while not exit:
    headers = [column['label'] for column in headers_columns]
    options = ''
    index = 1

    for column in headers:
      options += f'{column.replace('_', ' ').title()}, '
      index += 1
    options = options[0:len(options) - 2]
    print('-Buscador de alumnos')
    print('-Si deseas ir al menu principal, escribe "salir"')
    print(f'-Puedes consultar por cualquier dato:')
    print(f'-({options})')
    text = input('Buscar: ')
    if text == 'salir':
      exit = True
    else:
      search(dataBase, headers_columns, text)
```

```python
# Eliminar alumnos

def init_delete_record():
  print('Eliminar alumno del Master')
  controlNum = input('Ingresa el número de control: ')
  student = find_by(dataBase, 'control_num', controlNum)
  if not is_empty(student['record']):
    dataBase.pop(student['index'])
    print(f'"{controlNum}- {student['record']['first_name']} {student['record']['last_name']}" eliminado correctamente.')
  else:
    print(f'"{controlNum}" no encontrado.')
```

```python
# Crear nuevos alumnos

def ganerate_new_dict(_list, headers, uniqueCol = ''):
  result = {
    "status": "error",
    "message": "",
    "data": {}
  }

  if is_empty(_list):
    result['message'] = "La lista no puede ser vacía."
    return result
  if is_empty(headers):
    result['message'] = "La cabecera no puede ser vacía."
    return result
  if not isinstance(_list, list):
    result['message'] = "El 1° argumento debe de ser una lista."
    return result
  if not isinstance(headers, list):
    result['message'] = "El 2° argumento debe de ser una lista."
    return result

  index = 1
  newRow = {}

  for column in headers:
    col = column['col']
    label = column['label']
    formattedCol = label.title()
    inputTitle = ''
    newRow[col]= ''
    while is_empty(newRow[col]):
      try:
        if column['type'] == 'list':
          inputTitle = f'{index}. Ingresar ({formattedCol}) separados por comas: '
          inputText = input(inputTitle)
          newRow[col] = inputText.split(',')
        else:
          inputTitle = f'{index}. Ingresar ({formattedCol}): '
          newRow[col] = input(inputTitle)
        if is_empty(newRow[col]):
          print('No se aceptan vacíos.')
        if col == uniqueCol:
          foundRow = find_by(_list, uniqueCol, newRow[col])
          if not is_empty(foundRow['record']):
            print(f'{label} ya existe, ingresa uno nuevo.')
            newRow[col] = ''
        if newRow[col] == 'salir':
          result['status'] = 'exit'
          result['message'] = "Saliendo..."
          return result
      except:
        print('Dato incorrecto, vulve a intentar.')
    index += 1
  result['status'] = 'ok'
  result['message'] = "Registro generado con exito."
  result['data'] = newRow
  return result

def init_new_student():
  print('Nuevo alumno')
  print('-Todos los datos son obligatorios')
  print('-Si deseas ir al menu principal, escribe "salir"')
  result = ganerate_new_dict(dataBase, headers_columns, 'control_num')
  if result['status'] == 'ok':
    dataBase.append(result['data'])
    print(result['message'])
    view_table([result['data']], headers_columns)
  elif result['status'] == 'exit':
    print(result['message'])
  else:
    print(result['message'])
```

```python
# Actualizar alumnos

def update_dict(_list, headers, uniqueCol = ''):
  result = {
    "status": "error",
    "message": "",
    "data": {}
  }

  if is_empty(_list):
    result['message'] = "La lista no puede ser vacía."
    return result
  if is_empty(headers):
    result['message'] = "La cabecera no puede ser vacía."
    return result
  if not isinstance(_list, list):
    result['message'] = "El 1° argumento debe de ser una lista."
    return result
  if not isinstance(headers, list):
    result['message'] = "El 2° argumento debe de ser una lista."
    return result

  index = 1
  selectOption = {}
  newRow = {}
  menu = ''
  uniqueLabel = ''

  for column in headers:
    col = column['col']
    label = column['label']
    if uniqueCol == col:
      uniqueLabel = label
      continue
    formattedCol = label.title()
    menu += f'{index}. Editar ({formattedCol}).\n'
    selectOption[index] = { "label": label, "col": col }
    index += 1

  optionSelected = 1
  inputSearch = ''
  foundRecord = {}
  record = {}
  oldRow = {}
  listIndex = 0

  while (optionSelected != 'salir' and inputSearch != 'salir'):
    if is_empty(foundRecord):
      inputSearch = input(f'Ingresa ({uniqueLabel.lower()}): ')
      foundRecord = find_by(dataBase, uniqueCol, inputSearch)
      record = foundRecord['record']

    if inputSearch == 'salir':
      continue

    if is_empty(record):
      foundRecord = {}
      print('Registro no encontrado, intenta de nuevo.')
      continue

    listIndex = foundRecord['index']
    print('Registro seleccionado:')
    view_table([record], headers)
    print('Selecciona una opción del menu')
    print(menu)

    try:
      optionSelected = input('Ingresa una opción valida: ')
      if optionSelected == 'salir':
        continue
      optionSelected = int(optionSelected)
    except:
      print('Opción no valida, favor de seleccionar una valida.\n')
      continue

    if (is_empty(selectOption[optionSelected])):
      print('Opción incorrecta, intenta de nuevo.')
      continue

    col = selectOption[optionSelected]['col']
    label = selectOption[optionSelected]['label']

    if uniqueCol == col:
      print("La clave primaria no puede ser actualizada.")
      continue

    newValue = input(f'Ingresa el ({label}): ')

    oldRow = record
    newRow = oldRow
    newRow[col] = newValue

    result['status'] = 'ok'
    result['message'] = "Registro generado con exito."
    result['data'] = { "oldRow": oldRow, "newRow": newRow, "index": listIndex }
    inputSearch = 'salir'
    optionSelected = 'salir'

  if result['status'] == 'ok':
    return result
  else:
    result['status'] = 'exit'
    result['message'] = 'Saliendo..'
    return result

def init_update_student():
  print('Menu editar alumno')
  print('-Si deseas ir al menu principal, escribe "salir"')
  print('-Para editar, primero escribe el número de control\n')
  result = update_dict(dataBase, headers_columns, 'control_num')

  if result['status'] == 'ok':
    index = result['data']['index']
    newRow = result['data']['newRow']
    dataBase[index] = newRow
    print('El siguiente registro fue actualizado exitosamente:')
    view_table([newRow], headers_columns)
  else:
    print(result['message'])
```

```python
# Función que inicializa el menu de (Alumnos)

def executeStudentsCRUD():
  optionSelected = None
  while (optionSelected != 6):
    try:
      print('')
      show_students_menu()
      optionSelected = int(input('Ingresa una opción del menu: '))
      print('')

      if (optionSelected > 0 and optionSelected < 7):
        if (optionSelected == 1):
          view_table(dataBase, headers_columns)
        elif (optionSelected == 2):
          init_new_student()
        elif (optionSelected == 3):
          init_find_students()
        elif (optionSelected == 4):
          init_update_student()
        elif (optionSelected == 5):
          init_delete_record()
      else:
        print('Opción no valida, vuleve a intentar.\n')
    except Exception as e:
      print('Ocurrió un error.\n', e)
      optionSelected = None
  print('Saliendo del menu...')
```
